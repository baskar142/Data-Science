```python
def freqTable(columnName, df):
    freqTable = pd.DataFrame(columns=["Unique_values","Frequncy", "Relative_Frequncy", "Cumulative"])
    freqTable["Unique_values"] = df[columnName].value_counts().index
    freqTable["Frequncy"] = df[columnName].value_counts().values
    freqTable["Relative_Frequncy"] = (freqTable["Frequncy"]/100)
    freqTable["Cumulative"] = (freqTable["Relative_Frequncy"].cumsum())
    return freqTable
```

```python
def find_outlier(descriptive, quan):
    lesser = []
    greater = []     
          
    # Find columns with outliers
    for columnName in quan:
        if descriptive[columnName]["Min"] < descriptive[columnName]["Lesser"]:
            lesser.append(columnName)
        if descriptive[columnName]["Max"] > descriptive[columnName]["Greater"]:
            greater.append(columnName)
    return lesser, greater 
    
```

```python
lesser,greater = find_outlier(descriptive, quan)
print("lesser:",lesser)
print("greater:",greater)
```

```python
    # Replace outliers
def replace_outlier(descriptive, quan, df):    
    for columnName in lesser:
        df.loc[df[columnName] < descriptive[columnName]["Lesser"], columnName] = descriptive[columnName]["Lesser"]        
    for columnName in greater:
        df.loc[df[columnName] > descriptive[columnName]["Greater"], columnName] = descriptive[columnName]["Greater"]    
    return df
```
