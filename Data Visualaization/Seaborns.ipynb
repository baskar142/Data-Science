```python
df =sns.load_dataset('exercise')
sns.factorplot(x ='time', y='pulse', hue='kind', kind='violin', col='diet', data=df);
plt.show()
```

```python
df =sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=df)
sns.lmplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
df =sns.load_dataset('anscombe')
sns.lmplot(x='x', y='y', data=df)
plt.show()
```

```python
# Create the point plot
plt.figure(figsize=(12, 6))
sns.pointplot(x='specialisation', y='mba_p', hue='status', data=df, dodge=True, markers=['o', 's'], capsize=.1, errwidth=1, palette='colorblind')

# Customize the plot
plt.title('MBA Percentage by Specialisation and Status')
plt.xlabel('Specialisation')
plt.ylabel('MBA Percentage')
plt.legend(title='Status')
plt.show()
```
